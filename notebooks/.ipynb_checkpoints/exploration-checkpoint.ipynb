```python
import pandas as pd

pd.set_option('display.max_columns', None)

chunks = pd.read_json("../data/data.json", lines=True, chunksize=10000)



```

```python
i = 0
chunk_list = []
for chunk in chunks:
    chunk = chunk.loc[(chunk["following_list"].notna()) & (chunk["repo_list"].notna())]
    chunk = chunk.loc[(chunk["is_suspicious"] == False) 
                    & (chunk["type"] == "User") 
                    & (chunk["bio"].notna())
                    & (chunk["company"].notna())
                    & (chunk["following_list"].map(lambda d: len(d)) > 0)
                    & (chunk["repo_list"].map(lambda d: len(d)) > 0)
                ]

    df_subset = chunk[["id", "company", "bio", "follower_list", "following_list", "repo_list"]]
    df_subset
    chunk_list.append(df_subset)

df = pd.concat(chunk_list)

df.to_csv("../data/data.csv", encoding='utf-8', index=False)
df
```

```python
df = df.loc[(df["following_list"].notna()) & (df["repo_list"].notna())]
df = df.loc[(df["is_suspicious"] == False) 
                & (df["type"] == "User") 
                & (df["bio"].notna())
                & (df["company"].notna())
                & (df["following_list"].map(lambda d: len(d)) > 0)
                & (df["repo_list"].map(lambda d: len(d)) > 0)
            ]

df_subset = df[["id", "company", "bio", "follower_list", "following_list", "repo_list"]]
df_subset
```
