```python
import pandas as pd
chunks = pd.read_json("../data/data.json", lines=True, chunksize=10000)
```

# Filtering Data

```python
chunk_list = []
chunk_no = 0
for chunk in chunks:
   
    chunk = chunk.loc[(chunk["repo_list"].notna()) & (chunk["bio"].notna()) & (chunk["repo_list"].map(lambda d: len(d) if d != None else 0) > 0)
    & (chunk["is_suspicious"] == False) 
                    & (chunk["type"] == "User") ][["id", "location", "company", "bio", "follower_list", "following_list", "repo_list"]]

    chunk_list.append(chunk)
    chunk_no += 10000
    print(chunk_no)

df = pd.concat(chunk_list)

df.to_csv("../data/data_repo_and_bio.csv", encoding='utf-8')
```

```python
job_df = df.copy()
```

```python
job_df = job_df[["id", "bio", "repo_list"]]
```

```python
for i in job_df.index:
    repo_list = job_df.at[i, "repo_list"]
    languages = [d["language"] for d in repo_list]
    languages = [i for i in set(languages) if i is not None]
    job_df.at[i, 'repo_list'] = languages
job_df.rename(columns={"repo_list": "languages"}, inplace=True)
job_df = job_df.loc[(job_df["languages"].map(lambda d: len(d)) > 0)]
```

```python
job_df = job_df[["bio", "languages"]]
```

```python
job_df.to_csv("../data/data_bio_and_languages.csv", encoding='utf-8')
```
